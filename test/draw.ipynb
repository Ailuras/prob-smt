```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# 假设你已经有了这个列表
x_coords = np.random.uniform(-1, 1, 1000)
y_coords = np.random.uniform(-1, 1, 1000)
points = list(zip(x_coords, y_coords))

# 定义三个顶点的坐标
points_triangle = np.array([[-1, -1], [0, 0], [1, -1]])

# 创建一个新的figure对象
fig, ax = plt.subplots(figsize=(8, 8))

ax.set_aspect('equal', 'box')

# 创建一个多边形patch
polygon = patches.Polygon(points_triangle, closed=True, fill=True, color='blue', alpha=0.2) 

# 添加这个patch到你的axes对象
ax.add_patch(polygon)

# 绘制三角形的边
ax.plot(*points_triangle.T, color='blue', lw=0.5)

# 将100组坐标绘制到图中
for point in points:
    ax.plot(*point, 'ro', markersize=1)
    
plt.plot(0, -1/2, 'kd', markersize=15)

# 设置x和y轴的范围
plt.xlim(-1, 1)
plt.ylim(-1, 1)

# 添加网格线
ax.grid(True)

# 显示图形
plt.savefig('./charts/UD.pdf')
plt.show()
```

```python
from scipy.stats import norm

# 均值为0，标准差为1的标准正态分布
mu = 0
sigma = 1

# 想要获取概率密度函数值的点
x = 0

# 使用norm.pdf方法获取函数值
pdf_value = norm.pdf(x, mu, sigma)

print(f"The value of the probability density function at x = {x} is {pdf_value}")
```
